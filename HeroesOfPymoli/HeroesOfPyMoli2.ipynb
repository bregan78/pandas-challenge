```python
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

new_purchase_data = purchase_data.rename(columns={'Purchase ID': 'Purchase_ID',
                             'Item Name': 'Item_Name'})
new_purchase_data.head()
```

```python
unique_ID = len(new_purchase_data["Item ID"].unique().tolist())
row_count = len(new_purchase_data["Purchase_ID"])

sum_Price = new_purchase_data["Price"].sum()

avg_price =  sum_Price / row_count 
avg_price
unique_ID
new_purchase_data.Price.count()
new_purchase_data.Price.sum()

```

```python
no_dups = new_purchase_data.drop_duplicates(['SN'], keep='last')
no_dups

gender_df = no_dups.groupby("Gender")

male_df = gender_df.get_group("Male")
count_of_males = male_df.SN.count()

female_df = gender_df.get_group("Female")
count_of_females = female_df.SN.count()
other_df = gender_df.get_group("Other / Non-Disclosed")

count_of_others = other_df.SN.count()

total_count_no_dups = count_of_males + count_of_females + count_of_others
percent_of_males = (count_of_males/total_count_no_dups) * 100
percent_of_females = (count_of_females/total_count_no_dups) * 100
percent_of_others= (count_of_others/total_count_no_dups) * 100
percent_of_others

gender_new_df_data= {'Total Count': [count_of_males, count_of_females, count_of_others], 
        'Percentage of Players': [percent_of_males, percent_of_females,percent_of_others]}


gender_demo = pd.DataFrame(gender_new_df_data, index = ['Male', 'Female', 'Other / Non-Disclosed'])

gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map('{:.2f}%'.format)


gender_demo
```
